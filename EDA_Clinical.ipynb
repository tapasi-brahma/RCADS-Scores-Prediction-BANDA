**Converting RCADS string entries to numerical values:** 

```python
insert_index = 56

old_column_index = 1

for RCADS_params in range(9,56):
    
    RCADS_score = []
    
    for row in range(1,len(Clinical_Data)): #583 rows in Clinical_Data dataframe
    
        if Clinical_Data.iloc[:,RCADS_params][row] == 'Never':
        
            RCADS_score.append(0)
        
        elif Clinical_Data.iloc[:,RCADS_params][row] == 'Sometimes':
            
            RCADS_score.append(1)
            
        elif Clinical_Data.iloc[:,RCADS_params][row] == 'Often':
                
            RCADS_score.append(2)
                
        elif Clinical_Data.iloc[:,RCADS_params][row] == 'Always':
                    
            RCADS_score.append(3)
                    
        else:
            
            Clinical_Data.iloc[row,RCADS_params] = 'missing'
            RCADS_score.append('missing')
            
    
    RCADS_score.insert(0, Clinical_Data.iloc[:,RCADS_params][0])  
        
    #print(len(RCADS_score))
    
    Clinical_Data.insert(insert_index, Clinical_Data.columns[RCADS_params] + '_score', RCADS_score)
    
    insert_index = insert_index + 1
    
```
